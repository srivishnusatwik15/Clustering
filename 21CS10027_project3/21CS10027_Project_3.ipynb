```python
def cos_distance(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    return 1 - (dot_product / (norm_a * norm_b))

def assign_clusters(df, centroids):
    clusters = [[] for _ in range(len(centroids))]
    for point in df:
        distance = [cos_distance(point, centroid) for centroid in centroids]
        cluster_index = np.argmin(distance)
        clusters[cluster_index].append(point)
    return clusters

def update_centroids(clusters):
    return [np.mean(cluster, axis=0) for cluster in clusters]

def kmeans(df, k, iterations=20):
    centroids = df[np.random.choice(df.shape[0], k, replace=False)]
    for _ in range(iterations):
        clusters = assign_clusters(df, centroids)
        centroids = update_centroids(clusters)
    return clusters, centroids
```

```python
def cosine_distance(a,b):
    return 1 - cosine_similarity([a], [b])[0][0]

def pairwise_distances(df1, df2=None):
    if df2 is None:
        df2 = df1
    l1 = len(df1)
    l2 = len(df2)
    distance = np.zeros((l1, l2))
    for i in range(l1):
        for j in range(l2):
            distance[i][j] = cosine_distance(df1[i], df2[j])
    return distance

def mean_in_cluster_distance(df, labels, i):
    cluster_points = df[labels == i]
    if len(cluster_points) < 2:
        return 0 
    return np.mean(pairwise_distances(cluster_points))

def mean_nearest_cluster_distance(df, labels, i):
    cluster_points = df[labels == i]
    if len(cluster_points) == 0:
        return 0
    min_distance = float('inf')
    for j in range(len(labels)):
        if j != i:
            other_cluster_points = df[labels == j]
            if len(other_cluster_points) < 1:
                continue 
            distance = np.mean(pairwise_distances(cluster_points, other_cluster_points))
            if distance < min_distance:
                min_distance = distance
    return min_distance

def silhouette_coefficient_for_point(df, labels, i):
    a_i = mean_in_cluster_distance(df, labels, i)
    b_i = mean_nearest_cluster_distance(df, labels, i)
    if a_i == 0 and b_i == 0:
        return 0
    return (b_i - a_i) / max(a_i, b_i)

def silhouette_score(df, labels):
    coeff = [silhouette_coefficient_for_point(df, labels, i) for i in range(len(labels))]
    return np.mean(coeff)

def calculate_silhouette_coefficient(df, clusters):
    labels = np.zeros(len(df))
    for i, cluster in enumerate(clusters):
        labels[cluster] = i

    silhouette_avg = silhouette_score(df, labels)
    return silhouette_avg
```

```python
silhouette_coeff = calculate_silhouette_coefficient(np.vstack(clusters), [i for i in range(len(clusters))])
print("Silhouette Coefficient Value : ", silhouette_coeff)
```

```python
def find_optimal_k(df, max_k=6):
    optimal_k = None
    max_silhouette_coeff = -1
    optimal_clusters = None

    for k in range(3, max_k+1):
        clusters, _ = kmeans(df, k)
        silhouette_coeff = calculate_silhouette_coefficient(np.vstack(clusters), [i for i in range(len(clusters))])
        print("For k =", k, ", Silhouette Coefficient Value:", silhouette_coeff)

        if silhouette_coeff > max_silhouette_coeff:
            max_silhouette_coeff = silhouette_coeff
            optimal_k = k
            optimal_clusters = clusters

    return optimal_k, max_silhouette_coeff, optimal_clusters
```

```python
def divisive_clustering(df, k):
    C = []
    C0 = [i for i in range(len(df))] 
    C.append(C0)

    while len(C) < k:
        max_similarity = 2 
        max_index = 0
        print('Number of Clusters: '+str(len(C)))
        for i in range(len(C)):
            temp = C[i]
            if len(temp) == 1:
                continue
            for j in range(len(temp)):
                for l in range(j + 1, len(temp)):
                    similarity = np.dot(df[temp[j]], df[temp[l]])
                    similarity = similarity/ (np.linalg.norm(df[temp[j]]) * np.linalg.norm(df[temp[l]]) )
                    if similarity < max_similarity:
                        max_similarity = similarity
                        max_index = i

        Cs = C[max_index]
        print('Cluster to be divided: '+str(max_index))
        C.remove(Cs)
        temp = []
        for i in Cs:
            sim = 0
            for j in Cs:
                if i == j:
                    continue
                sim += np.arccos(np.dot(df[i], df[j])/(np.linalg.norm(df[i]) * np.linalg.norm(df[j])))
            if len(Cs) > 1:
                sim = sim / (len(Cs) - 1)
            temp.append(sim)

        C1 = [Cs[np.argmax(temp)]]
        Cs.remove(Cs[np.argmax(temp)])

        while len(Cs) > 0:
            print('Number of elements in Cluster: '+str(len(Cs)))
            final_sims = []
            for i in Cs:
                intra_sim = 0
                for j in Cs:
                    if i == j:
                        continue
                    intra_sim += np.arccos(np.dot(df[i], df[j])/(np.linalg.norm(df[i]) * np.linalg.norm(df[j])))
                intra_sim = intra_sim / (len(Cs) - 1) if len(Cs) > 1 else 0
                inter_sim = 0
                for j in C1:
                    inter_sim += np.arccos(np.dot(df[i], df[j])/(np.linalg.norm(df[i]) * np.linalg.norm(df[j])))
                inter_sim = inter_sim / len(C1)
                final_sims.append(intra_sim - inter_sim)

            if max(final_sims) < 0:
                break
            C1.append(Cs[np.argmax(final_sims)])
            Cs.remove(Cs[np.argmax(final_sims)])

        C.append(C1)
        C.append(Cs)
        print('Number of Clusters: '+str(len(C)))

    return C
```

```python
k = optimal_k
clusters = divisive_clustering(data_arr, k)

with open('divisive_clusters_info.txt', 'w') as file:
    file.write(f'Number of Clusters: {len(clusters)}\n\n')
    for cluster_idx, cluster in enumerate(clusters):
        file.write(f'Cluster {cluster_idx + 1}:\n')
        for point_idx in cluster:
            file.write(' '.join(map(str, data_arr[point_idx])) + '\n')
        file.write('\n')
```
